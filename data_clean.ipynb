```python
import os
```

```python
# Base directory where all city folders are present
base_dir = r"G:\DataScience_projects\Text2Sql_chatbot\archive"

# Loop through each folder (city name)
for city in os.listdir(base_dir):
    city_path = os.path.join(base_dir, city)
    
    # Check if it's a folder (skip files directly inside archive)
    if os.path.isdir(city_path):
        all_dfs = []
        
        # Loop through all CSV files inside this city folder
        for file in os.listdir(city_path):
            if file.endswith(".csv"):
                file_path = os.path.join(city_path, file)
                try:
                    df = pd.read_csv(file_path, sep="|")
                    all_dfs.append(df)
                except Exception as e:
                    print(f" Error reading {file_path}: {e}")
        
        if all_dfs:
            # Merge all dataframes
            merged_df = pd.concat(all_dfs, ignore_index=True)
            
            # Save merged CSV in the base_dir with city name
            output_file = os.path.join(base_dir, f"{city}.csv")
            merged_df.to_csv(output_file, index=False, sep="|")
            print(f" Merged CSV saved: {output_file}")
        else:
            print(f" No CSV files found in {city_path}")
```

```python
# Path to the 'all' folder
all_folder = r"G:\DataScience_projects\Text2Sql_chatbot\archive\all"

# List to store all DataFrames
all_dfs = []

# Loop through all CSV files in the 'all' folder
for file in os.listdir(all_folder):
    if file.endswith(".csv"):
        file_path = os.path.join(all_folder, file)
        try:
            df = pd.read_csv(file_path, sep="|")
            all_dfs.append(df)
        except Exception as e:
            print(f"Error reading {file_path}: {e}")

# Merge all DataFrames if any exist
if all_dfs:
    merged_df = pd.concat(all_dfs, ignore_index=True)
    # Save the merged CSV in the archive folder as all.csv
    output_file = r"G:\DataScience_projects\Text2Sql_chatbot\archive\all_places.csv"
    merged_df.to_csv(output_file, index=False, sep="|")
    print(f"Merged CSV saved as: {output_file}")
else:
    print("No CSV files found in the 'all' folder.")
```
